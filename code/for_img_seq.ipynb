```python
file = "1849"
videoPath = file+'/'+file+'.mp4'
print(videoPath)

vidcap = cv2.VideoCapture(file+'/'+file+'.mp4')
success,image = vidcap.read()
count = 1
while success:
  #image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
  ret,th = cv2.threshold(image,5,255,cv2.THRESH_BINARY)
  cv2.imwrite(file+"/"+file+"_%d.png" % count, th)     # save frame as PNG file      
  success,image = vidcap.read()
  print('Read a new frame: ', success)
  count += 1
```
